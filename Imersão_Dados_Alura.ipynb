# Aula 4 - Prevendo notas com Machine Learning

Será usado as notas das demais provas no modelo de machine learning para "prever" a nota da prova de Matemática.

```
provas_entrada = ["NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_CN", "NU_NOTA_REDACAO"]
provas_saida = "NU_NOTA_MT"

dados_sem_nota_zero = dados_sem_nota_zero[provas].dropna() # retira os NaN

notas_entrada = dados_sem_nota_zero[provas_entrada]
notas_saida = dados_sem_nota_zero[provas_saida]

notas_entrada
```

```
x = notas_entrada
y = notas_saida
# usualmente essa é a nomenclatura para os dados de entrada (x)
# e saída (y)
```

```
from sklearn.model_selection import train_test_split

SEED = 4321 # esse parâmetro é passado para evitar a randomização dos resultados

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25,
                                                        random_state=SEED)
# ao executar o train_test_split, ele gera dados de treino e teste tanto
# para os dados de entrada como para os de saída
```

```
x_treino.head()
```

As linhas das variáveis de entrada e saída, tanto de treino quanto teste, precisam ser iguais.

```
x_teste.shape 
```

```
len(y_teste)
```

Treino usando o modelo de regressão linear do tipo SVR (Linear Support Vector Regression)

```
from sklearn.svm import LinearSVR

# passando o SEED evita a aleatoriadade dos resultados
modelo = LinearSVR(random_state=SEED)
modelo.fit(x_treino, y_treino)
```

A seguir são feitas as predições das notas de Matemática usando as notas das demais matérias.

```
predicoes_matematica = modelo.predict(x_teste)
```

```
y_teste[:5] # resultado aproximado obtido pela regressão
```

```
plt.figure(figsize=(8, 8))
plt.grid()
sns.scatterplot(x=predicoes_matematica, y=y_teste)
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

Segundo o gráfico, a predição tendeu muito o erro para mais ou para menos na medida em que a nota da prova aumentava.

```
plt.figure(figsize=(8, 8))
plt.grid()
sns.scatterplot(x=y_teste, y=y_teste - predicoes_matematica)
# visualização do erro entre os dados reais e a predição
```

Aqui fica claro que houve um erro considerável entre as notas reais e as da predição. Quanto mais próximo de uma reta, melhor seria a predição.

Usando a validação cruzada (`cross_validation`), é possível dividir em n-partes a amostra de dados e assim poder treinar e validar várias vezes diferentes amostras.
Por exemplo: numa amostra de 100 dados, divide-se em 5 partes (que é a divisão padrão desse modelo) de 20 dados cada, 20 desses dados são usados para treino, enquanto os demais 80 seriam usados para a validação. E assim cada parte seria utilizada para treino e em conjunto para a validação.

```
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
# a saída dessa função é um dicionário
resultado = cross_validate(modelo_arvore, x, y, scoring="neg_mean_squared_error")
resultado["test_score"]*(-1) # isso é feito para deixar positivo o resultado
```

```
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
# o parâmetro cv é quem diz quantas divisões de amostras serão feitas
resultado = cross_validate(modelo_arvore, x, y, cv=10, scoring="neg_mean_squared_error")
resultado["test_score"]*(-1)
```

Como a média do erro não é uma maneira que representa tão fielmente o resultado de um modelo, usa-se o desvio-padrão para checar melhor o resultado.

```
desvio_padrao = (resultado["test_score"]*(-1)).std()
lim_inferior = media - (2 * desvio_padrao)
lim_superior = media + (2 * desvio_padrao)

print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

```
def calcula_mse(resultado):
    media = (resultado["test_score"]*-1).mean()
    desvio_padrao = (resultado["test_score"]*-1).std()
    lim_inferior = media - (2*desvio_padrao)
    lim_superior = media + (2*desvio_padrao)
    print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

A função Kfold embaralha a base de dados usadas no modelo pra que isso evite erros que possam vir de dados ordenados numericamente (crescente ou decrescente), alfabeticamente etc.

```
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

# isso fixa os resultados obtidos nessa célula
SEED = 1232
np.random.seed(SEED)

# divide de forma aleatória as amostras
partes = KFold(n_splits=10, shuffle=True) 

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultado = cross_validate(modelo_arvore, x, y, cv=partes, scoring="neg_mean_squared_error")
calcula_mse(resultado)
```

```
def regressor_arvore(nivel):
    SEED= 1232
    np.random.seed(SEED)
    partes = KFold(n_splits=10, shuffle=True)
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultados = cross_validate(modelo_arvore, x, y, cv=partes, scoring="neg_mean_squared_error", return_train_score=True)
    print(f"Treino = {(resultados['train_score']*-1).mean()} | Teste = {(resultados['test_score']*-1).mean()}")

regressor_arvore(4)
```

Checagem se quanto mais profundo for a árvore do modelo (`max_depth`), melhor seria os resultados.

```
for i in range(1,21):
    regressor_arvore(i)
```

Nota-se que na medida que os resultados do treino diminuem, o que é algo bom, os resultados de teste pioram, aumentando o seu valor, o que mostra que durante o teste o modelo não consegue generalizar mais os dados utilizados e daí ele meio que ficou "viciado" com o treino, ou seja, o modelo decorou os dados e assim não é capaz de fazer a predição de forma correta. Isso é chamado de **overfit**.

Por isso, nem sempre utilizar um modelo de árvore de decisão com uma profundidade muito grande pode resultar em menos erro.

**Desafio 4: Pesquisar o que é o problema de underfit**

*Overfitting*

Um cenário de overfitting ocorre quando, nos dados de treino, o modelo tem um desempenho excelente, porém quando utilizamos os dados de teste o resultado é ruim. Podemos entender que, neste caso, o modelo aprendeu tão bem as relações existentes no treino, que acabou apenas decorando o que deveria ser feito, e ao receber as informações das variáveis preditoras nos dados de teste, o modelo tenta aplicar as mesmas regras decoradas, porém com dados diferentes esta regra não tem validade, e o desempenho é afetado. É comum ouvirmos que neste cenário o modelo treinado não tem capacidade de generalização.

*Underfitting*

Neste cenário o desempenho do modelo já é ruim no próprio treinamento. O modelo não consegue encontrar relações entre as variáveis e o teste nem precisa acontecer. Este modelo já pode ser descartado, pois não terá utilidade.

**Desafio 5: Plotar um gráfico com test_score e train_test**

```
def regressor_arvore_plot(nivel):
    SEED= 1232
    np.random.seed(SEED)
    partes = KFold(n_splits=10, shuffle=True)
    modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
    resultados = cross_validate(modelo_arvore, x, y, cv=partes, scoring="neg_mean_squared_error", return_train_score=True)
    treino = (resultados['train_score']*-1).mean()
    teste = (resultados['test_score']*-1).mean()
    return (treino, teste)

dados_treino = []
dados_teste = []

for i in range(1,21):
    a, b = regressor_arvore_plot(i)
    dados_treino.append(a)
    dados_teste.append(b)
```
